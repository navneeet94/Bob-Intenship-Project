```python
sales_person_target = pd.read_csv(url)
sales_person_target
```

```python
dummy_airtable_unique_year = dummy_airtable_df['Getting Year'].unique().tolist()
salesPerson_unique_year = sales_person_target['Year'].unique().tolist()
print(dummy_airtable_unique_year)
print(salesPerson_unique_year)
print()

dummy_airtable_unique_quarter = dummy_airtable_df['Getting Quarter'].unique().tolist()
salesPerson_unique_quarter = sales_person_target['Quarter'].unique().tolist()
print(dummy_airtable_unique_quarter)
print(salesPerson_unique_quarter)
print()


dummy_airtable_unique_salesperson = dummy_airtable_df['Sales Person'].unique().tolist()
salesPerson_person = sales_person_target['Sales Person'].unique().tolist()

print(f"{dummy_airtable_unique_salesperson}, Length = {len(dummy_airtable_unique_salesperson)}")
print(f"{salesPerson_person}, Length = {len(salesPerson_person)}")
```

```python
sprints = []
for airtable_year in dummy_airtable_unique_year:
    # Finding year is same or not
    for salesperson_year in salesPerson_unique_year:
        if airtable_year == salesperson_year:
            # Finding quarter is same or not
            for airtable_quarter in dummy_airtable_unique_quarter:
                for salesperson_quarter in salesPerson_unique_quarter:
                    if airtable_quarter == salesperson_quarter:
                        # Finding sales person is same or not
                        
                        for airtable_person in dummy_airtable_unique_salesperson:
                            if airtable_person in salesPerson_person:
                                for salesperson_val in salesPerson_person:                                
                                    if airtable_person == salesperson_val:
                                        print(f"Year :- {airtable_year} , Quarter :- {airtable_quarter}")
                                        print(airtable_person, end=" == ")
                                        print(salesperson_val, end=f", Target Amount {sales_person_target[(sales_person_target['Sales Person'] == salesperson_val)]['Target Amount'].values[0]}")
                                        print()
                                        person_founded = dummy_airtable_df[(dummy_airtable_df['Getting Year'] == airtable_year) & (dummy_airtable_df['Getting Quarter'] == airtable_quarter) & (dummy_airtable_df['Sales Person'] == salesperson_val)]
                                        #print(len(person_founded))
                                        person_len = len(person_founded)
                                        counter = 0
                                        if person_len > 0:
                                            person_founded['Target Amount'] = sales_person_target[(sales_person_target['Sales Person'] == salesperson_val)]['Target Amount'].values[0]
                                            print()
                                            if person_len > 1:
                                                person_founded[1:]["Target Amount"] = 0
                                            sprints.append(person_founded)
                            else:
                                print(airtable_person)
                                person_not_available = dummy_airtable_df[(dummy_airtable_df['Getting Year'] == airtable_year) & (dummy_airtable_df['Getting Quarter'] == airtable_quarter) & (dummy_airtable_df['Sales Person'] == airtable_person)]
                                person_not_available["Target Amount"] = 0
                                sprints.append(person_not_available)
                    else:
                        quarter_not_matched = dummy_airtable_df[(dummy_airtable_df['Getting Year'] == airtable_year) & (dummy_airtable_df['Getting Quarter'] == airtable_quarter)]
                        quarter_not_matched['Target Amount'] = 0
                        sprints.append(quarter_not_matched)
        else:
            year_not_matched = dummy_airtable_df[(dummy_airtable_df['Getting Year'] == airtable_year)]
            year_not_matched['Target Amount'] = 0
            sprints.append(year_not_matched)
```

```python
sprintCounter = 0
columns = []
for i in sprints:
    if sprintCounter < 1:
        #print(i.keys().tolist())
        columns = i.keys().tolist()
    sprintCounter += 1

# columns dictonary create
columns
columns_dict = {}

for i in columns:
    print(i)
    columns_dict[i] = []
    
columns_dict
```

```python
data_counter = 0
for i in columns_dict:
    for j in range(0,len(sprints)):
        its_data = sprints[j][i]
        #print(sprints[j][i])
        for k in its_data:
            print(k)
            columns_dict[i].append(k)
    data_counter +=1
    print()
    print('--------------------------------------')
    print()
```
